```python
import cv2
import numpy as np
from pygame import mixer
from keras.models import load_model
import time

# Initialize the mixer
mixer.init()

# Load the sound file
sound_path = r'C:\Users\HP\Desktop\Drowsiness Detection\mixkit-alert-alarm-1005.wav'
try:
    sound = mixer.Sound(sound_path)
except FileNotFoundError:
    print(f"Sound file not found at: {sound_path}")
    sound = None

# Load Haar Cascade for face and eye detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')

# Load the pre-trained model
try:
    model = load_model('path_to_your_model.h5')  # Update with the correct model path
except Exception as e:
    print(f"Error loading model: {e}")
    model = None
# Start video capture
cap = cv2.VideoCapture(0)
Score = 0
last_alert_time = time.time()
cooldown_period = 10  # Cooldown period in seconds

while True:
    ret, frame = cap.read()
    if not ret:
        break

    height, width = frame.shape[0:2]
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.2, minNeighbors=3)
    
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, pt1=(x, y), pt2=(x+w, y+h), color=(255, 0, 0), thickness=3)
        roi_gray = gray[y:y+h, x:x+w]
        eyes = eye_cascade.detectMultiScale(roi_gray)

        for (ex, ey, ew, eh) in eyes:
            eye = frame[y+ey:y+ey+eh, x+ex:x+ex+ew]
            eye = cv2.resize(eye, (80, 80))
            eye = eye / 255.0
            eye = np.expand_dims(eye, axis=0)

            if model:
                prediction = model.predict(eye)

                if prediction[0][0] > 0.35:
                    cv2.putText(frame, 'closed', (10, height-20), fontFace=cv2.FONT_HERSHEY_COMPLEX_SMALL, fontScale=1, color=(255, 255, 255),
                                thickness=1, lineType=cv2.LINE_AA)
                    Score += 1
                    if Score > 6 and sound:
                        current_time = time.time()
                        if current_time - last_alert_time > cooldown_period:
                            try:
                                sound.play()
                                last_alert_time = current_time  # Update last alert time
                            except Exception as e:
                                print(f"Error playing sound: {e}")
                elif prediction[0][1] > 0.85:
                    cv2.putText(frame, 'open', (10, height-20), fontFace=cv2.FONT_HERSHEY_COMPLEX_SMALL, fontScale=1, color=(255, 255, 255),
                                thickness=1, lineType=cv2.LINE_AA)
                    Score -= 1
                    if Score < 0:
                        Score = 0

                cv2.putText(frame, 'Score ' + str(Score), (100, height-20), fontFace=cv2.FONT_HERSHEY_COMPLEX_SMALL, fontScale=1, color=(255, 255, 255),
                            thickness=1, lineType=cv2.LINE_AA)

    cv2.imshow('frame', frame)
    if cv2.waitKey(33) & 0xFF == ord('q'):
        break

# Release resources
cap.release()
cv2.destroyAllWindows()
```
